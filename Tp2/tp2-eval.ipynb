```python
import numpy as np
import matplotlib.pyplot as plt

N=[1,1,20]
D=0.3324
Dmax=0.32832
z=45/60
X= lambda N: N/((N*D)+Z)
y=  lambda N: N/(D+z)
Z= 1/Dmax
d= lambda N: N
H= lambda N: N*D
def Zn(N):
    B=[]
    for i in N:
        B.append(1/Dmax)
    return B
T=np.linspace(1,20)
plt.title("Courbe du débit du système X(N)")
plt.plot(T, y(T), label="N/(D+z)")
plt.plot(T,X(T), label="N/((N*D)+Z)")
plt.plot(T, Zn(T),label="Z")
plt.legend()
plt.grid(True)
plt.show()
def Dn(N):
    D=[]
    for i in N:
        D.append((i*Dmax)-Z)
    return D

def En(N):
    E=[]
    for i in N:
        E.append(D)
    return E

plt.title("Courbe du temps de réponse R(N)")
plt.plot(T, H(T), label="ND")
plt.plot(T,Dn(T), label="NDmax")
plt.plot( T,En(T),label="D")
plt.legend()
plt.grid(True)
plt.show()
```
